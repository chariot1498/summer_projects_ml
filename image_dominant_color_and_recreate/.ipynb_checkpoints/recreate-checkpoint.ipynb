```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import cv2
from sklearn.cluster import KMeans
```

```python
aim = cv2.imread('./image.jpg')
im = cv2.cvtColor(aim, cv2.COLOR_BGR2RGB)
print im.shape
r, c = im.shape[:2]
out_r = 120
im = cv2.resize(im, (int(out_r*float(c)/r), out_r))
print im.shape
pixels = im.reshape((-1, 3))

print pixels.shape
plt.imshow(im)
```

```python
km = KMeans(n_clusters=30)
km.fit(pixels)
labels = km.labels_
centers = km.cluster_centers_
```

```python
labels.shape
type(centers)
centers.shape
centers[0][0]
```

```python
res = []
for ix in labels:
    i = list(centers[ix])
    res.append(i)
```

```python
res = np.array(res)
```

```python
res.resize(im.shape)
```

```python
res.shape
res = np.asarray(res,dtype=np.uint8)
```

```python
plt.imshow(res)
```
