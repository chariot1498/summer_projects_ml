```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import random
```

```python
# Define some initial params
# Dimensionality
Dn = 20

# Mutation probability
CR = 0.5

# Population size
NP = 200

# No. of iterations
I = 300

# Differential Weight
F = 0.48
```

```python
#We will try to optimize function sum over all dimensions ((x+r)^2) it will be our f(x) Where r is a random dn length
#vector
r = np.random.random((Dn,))
def f(x):
    return ((x + r)**2).sum()

print r
```

```python
# Generate population
# We are going to randomly Generate population that is NP points each with dn features from a uniform sample from -10 
# to 10.
pop = []

for ix in range(NP):
    # generate a random vector
    vec = np.random.uniform(-10, 10, (Dn,))
    
    # Add to population
    pop.append(vec)
#pop now has our initial random population
print len(pop)
```

```python
#In this for each iteration we first choose X_l which is the leader point this is the point with best fitness score
#i.e. least f(x) value for every point in the population we take random point in the pop and create a new vector y
#according to the F (differential score) now we randomly choose one dimension and check for that dimension if a randomly
#generated number has less prob than CR (mutation probability) if yes than for that dimension y[that dimension] is taken
#else the old value is taken for it , for other dimensions old value is taken . if the fitness score of this vector t
#is less than px the old vector then t is taken and old point is discarded .
loss = []
# Main DE loop
for ix in range(I):
    pop = sorted(pop, key=lambda z: f(z))
    print "Iteration: {0} | Best Value: {1}".format(ix, f(pop[0]))
    loss.append(f(pop[0]))
    
    # Leader point
    X_l = pop[0]
    
    for px_index in range(NP):
        px = pop[px_index]
        
        # pick a random point
        a,b,c = random.sample(pop, 3)
        
        # generate linear combination
        y = px + F*(X_l - c)
        
        # pick a random dimension
        R = random.choice(range(Dn))
        
        # temporary vector
        t = np.zeros(px.shape)
        
        for dx in range(Dn):
            r = np.random.random()
            if r < CR or dx == R:
                t[dx] = y[dx]
            else:
                t[dx] = px[dx]
        
        # check if new point is better
        if f(t) < f(px):
            pop[px_index] = t
        else:
            pass
```

```python
#best fitness scores through different iterations is plotted
plt. plot(loss)
```
