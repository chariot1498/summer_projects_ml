```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import cv2
from sklearn.cluster import KMeans
```

```python
aim = cv2.imread('./image.jpg')
im = cv2.cvtColor(aim, cv2.COLOR_BGR2RGB)
print im.shape
r, c = im.shape[:2]
out_r = 168
im = cv2.resize(im, (int(out_r*float(c)/r), out_r))
print im.shape
pixels = im.reshape((-1, 3))

print pixels.shape
plt.imshow(im)
```

```python
km = KMeans(n_clusters=10)
km.fit(pixels)
```

```python
im.shape
list(im)
```

```python
ii = []
for ix in range(r):
    for cx in range(c):
        l = list(im[ix][cx])
        l.append(ix)
        l.append(cx)
        ii.append(l)
ii = np.array(ii)
ii
```

```python
pixels = ii.reshape(-1,5)
```

```python
pixels
```

```python
km = KMeans(n_clusters=10)
km.fit(pixels)
```

```python
colors = np.asarray(km.cluster_centers_, dtype='uint8')
```

```python
colors.shape
```

```python
colors
```

```python
per = np.asarray(np.unique(km.labels_, return_counts=True)[1], dtype='float32')
per = per / pixels.shape[0]
per
```

```python
plt.figure(0)
for ix in range(colors.shape[0]):
    patch = np.ones((20, 20, 5))
    patch[:, :, :] = 255 - colors[ix]
    plt.subplot(1, colors.shape[0], ix+1)
    plt.axis("off")
    plt.imshow(patch)
```

```python
dom = [[per[ix], colors[ix]] for ix in range(km.n_clusters)]
DOM = sorted(dom, key=lambda z: z[0], reverse=True)
```

```python
plt.figure(0)
plt.axis('off')

patch = np.zeros((50, 500, 5))

start = 0
for cx in range(km.n_clusters):
    width = int(DOM[cx][0] * patch.shape[1])
    end = start + width
    patch[:, start:end, :] = 255 - DOM[cx][1]
    start = end

plt.imshow(patch)
plt.show()
```
